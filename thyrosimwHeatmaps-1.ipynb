```python
import numpy as np
import pandas as pd

temp = np.genfromtxt('parameter_variations.txt', delimiter=',') #avgT4data
df4 = pd.DataFrame(temp,columns=range(1,49))
df4=df4/78.334
df4=df4.apply(np.log)
df4 = df4.values 

my_dataT3 = np.genfromtxt('parameter_variationsT3.txt', delimiter=',') #avgT3data
df3 = pd.DataFrame(my_dataT3,columns=range(1,49))

df3=df3/1.3451
df3=df3.apply(np.log)
df33 = df3.values


my_dataTSH = np.genfromtxt('parameter_variationsTSH.txt', delimiter=',') #avgTSHdata
#print my_data, type(my_data)

dfTSH= pd.DataFrame(my_dataTSH,columns=range(1,49))
dfTSH=dfTSH/1.9918
dfTSH= dfTSH.apply(np.log)
dfTSH= dfTSH.values

#df4 = pd.DataFrame(my_dataT4,columns=range(1,49))
#print df
#for each element, divide by actual value and take log of it #t4-78.3


param_list=['S4','SR3/SR4 TimeDelay','K12','K13','K31Free','K21Free','A_Auxiliary','B_Auxiliary',' C_Auxiliary', 'D_Auxiliary','k4 T4absorption','k02','VmaxD1fast','KmD1fast','VmaxD1slow','KmD1slow','VmaxD2slow','KmD2slow','S3_Secretion Response to TSH','k45','k46','k64free','k54free','a_freeBindingConstant','b_freeBindingConstant', 'c_freeBindingConstant','d_freeBindingConstant','k3absorb', 'k05','Bzero','Azero','Amax','phi','kdegTSH-HYPO','VmaxTSH','K50TSH','k3','T4P-EU','T3P-EU','KdegT3B','KLAG-HYPO','KLAG','k4dissolve','k4excrete','k3dissolve','k3excrete','Vp_linearizationC','VTSH-linearizationC']
#names of all parameters in a list for labeling
step_axis= [-1,-0.8,-0.6,-0.4,-0.2,0,0.2,0.4,0.6,0.8,1]
# this is select rows 0 to 2 df[0:3] 

plt.close()
```

```python
#T4
vp = sns.heatmap(data=df4[:,0 :24], center=0,cmap="RdBu",cbar_kws={'label':'Effect on AvgT4 Secretion'}) 

vp.set_xticklabels(param_list[0:24])                            
plt.title('Effect of parameter variation from 0.1-10x on average T4 levels')
fig=plt.gcf()
fig.set_size_inches(12,5)
plt.ylabel("Multipled Quantity")
plt.xlabel("Parameters")
vp.set_yticklabels(step_axis)

plt.show()

vp1=sns.heatmap(data=df4[:,24 :49],center=0, cmap="RdBu",cbar_kws={'label':'Effect on AvgT4 Secretion'})
vp1.set_xticklabels(param_list[24:48])
fig1=plt.gcf()
fig1.set_size_inches(12,5)
plt.title('Effect of parameter variation from 0.1-10x on average T4 levels')
plt.xlabel("Parameters")
plt.ylabel("Multipled Quantity")
vp1.set_yticklabels(step_axis)

plt.show()
plt.close()
```
