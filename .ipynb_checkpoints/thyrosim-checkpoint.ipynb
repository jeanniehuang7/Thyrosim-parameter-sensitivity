```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# read in matrix from a csv file into a numpy array. format is (# of increments between 0.1 and 10) x (48 parameters)
my_data = np.genfromtxt('parameter_variations.txt', delimiter=',')

#make a dataframe based on my numpy array
df = pd.DataFrame(my_data,columns=range(1,49))
print df 
```

```python
# Example:
# For a sample dataset, make a violin plot
df = sns.load_dataset('iris')
# Make boxplot for one group only
sns.violinplot( y=df["sepal_length"] )
plt.show()
plt.close()
```

```python
vp = sns.violinplot(data=my_data[:,0:8])
plt.show()
plt.close()
```

```python
vp = sns.violinplot(data=my_data[:,0:24])
plt.title('Effect of parameter variation from 0.1-10x on average T4 levels')
fig=plt.gcf()
fig.set_size_inches(12,5)
plt.show()
plt.close()
```

```python
vp = sns.violinplot(data=my_data[:,24:46])
plt.title('Effect of parameter variation from 0.1-10x on average T4 levels')
fig=plt.gcf()
fig.set_size_inches(12,4)

plt.show()
plt.close()
```
